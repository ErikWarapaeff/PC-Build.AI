````python
import pandas as pd
import ast

def parse_list_columns(df: pd.DataFrame):
    # Преобразуем строки, представляющие списки, в настоящие списки, если это возможно
    def safe_parse_list(value):
        try:
            return ast.literal_eval(value) if isinstance(value, str) else value
        except (ValueError, SyntaxError):
            return value

    df['speed'] = df['speed'].apply(safe_parse_list)
    df['modules'] = df['modules'].apply(safe_parse_list)
    
    # Проверяем, что значения являются списками, и распаковываем их в новые колонки
    if isinstance(df['speed'].iloc[0], list):
        df[['speed_num', 'speed_value']] = pd.DataFrame(df['speed'].to_list(), index=df.index)
    if isinstance(df['modules'].iloc[0], list):
        df[['modules_num', 'modules_value']] = pd.DataFrame(df['modules'].to_list(), index=df.index)
    
    # Удаляем исходные колонки 'speed' и 'modules'
    df = df.drop(columns=['speed', 'modules'])
    
    return df

# Применяем функцию
parsed_data = parse_list_columns(df_2)
````

````python
from pydantic import BaseModel
from typing import List, Dict, Optional
import math
from datetime import datetime

# Define the system prompt
from langchain.prompts import ChatPromptTemplate

class Component(BaseModel):
    name: str
    price: float
    additional_specs: Dict[str, str]

class BuildResponse(BaseModel):
    type: str
    budget: float
    components: Dict[str, List[Component]]

def calculate_budget_distribution(budget: float, build_type: str, components_percentages: Dict[str, Dict[str, float]]) -> Dict[str, float]:
    """Distributes the budget across components based on the build type."""
    if build_type not in components_percentages:
        raise ValueError(f"Unsupported build type: {build_type}")
    
    distribution = components_percentages[build_type]
    return {component: round(budget * percentage, 2) for component, percentage in distribution.items()}

def fetch_components(agent_executor, component: str, budget: float, filters: Optional[Dict[str, str]] = None) -> List[Component]:
    """Performs SQL query via agent and returns results for a specific component."""
    
    # Define the system prompt context
    system_prompt = ( """
            <system_prompt>
            YOU ARE A SQL ASSISTANT AND EXPERT COMPONENT SELECTOR TASKED WITH HELPING USERS BUILD PC CONFIGURATIONS. YOUR PRIMARY OBJECTIVE IS TO SELECT COMPONENTS STEP-BY-STEP FROM DATABASE TABLES WHILE ENSURING COMPATIBILITY, ADHERING TO USER-DEFINED BUDGET CONSTRAINTS, AND ACCOUNTING FOR ANY ADDITIONAL USER REQUIREMENTS. YOU MUST FOLLOW THESE INSTRUCTIONS WITH PRECISION.

            ### INSTRUCTIONS ###

            - FOLLOW A STEP-BY-STEP PROCESS TO SELECT EACH COMPONENT OF A PC.
            - ENSURE EVERY COMPONENT IS COMPATIBLE WITH THE OTHERS BY REFERENCING COMPATIBILITY TABLES.
            - ADHERE TO THE USER'S BUDGET CONSTRAINTS AT EVERY STEP.
            - TAKE INTO ACCOUNT ADDITIONAL USER REQUIREMENTS, SUCH AS MINIMUM COMPONENT SPECIFICATIONS (E.G., VIDEO MEMORY, CORE COUNT, OR ADVANCED FEATURES LIKE 4K GAMING SUPPORT).
            - SAVE THE DETAILS OF EACH SELECTED COMPONENT FOR OUTPUT.
            - USE CLEAR, OPTIMIZED SQL QUERIES TO RETRIEVE AND FILTER DATA.

            ### ADDITIONAL USER REQUIREMENTS ###
            WHEN THE USER SPECIFIES ADDITIONAL REQUIREMENTS, SUCH AS:
            1. **MINIMUM VIDEO MEMORY (E.G., GREATER THAN 12GB)**:
            - ENSURE THE SELECTED GPU MEETS OR EXCEEDS THIS CRITERION BY CHECKING THE `memory` COLUMN IN THE `gpu_full_info` TABLE.
            2. **MINIMUM CPU CORE COUNT (E.G., GREATER THAN 6 CORES)**:
            - ENSURE THE SELECTED CPU MEETS THIS CRITERION BY CHECKING THE `core_count` COLUMN IN THE `cpu_merged` TABLE.
            3. **4K GAMING SUPPORT FOR THE GPU**:
            - VERIFY SUPPORT FOR 4K GAMING BY CHECKING THE `4K_Ultra` COLUMN IN THE `gpu_hierarchy` TABLE (E.G., PERFORMANCE RATINGS LIKE `100.0% (114.5fps)`).

            ### STEP-BY-STEP GUIDE ###

            #### STEP 1: SELECT A GPU
            1. QUERY THE `gpu_full_info` TABLE TO RETRIEVE GPU OPTIONS.
            2. FILTER RESULTS BASED ON:
            - THE USER'S BUDGET USING `average_price <= user_budget`.
            - CHOOSE ONLY UNIQUE GPUS OR FOLLOW THE INSTRUCTIONS
            - ADDITIONAL REQUIREMENTS SUCH AS:
                - VIDEO MEMORY (`memory >= {required_memory}`).
                - 4K GAMING SUPPORT (`4K_Ultra IS NOT NULL AND 4K_Ultra LIKE '%fps%'`).
                - ETC.
            3. RANK THE GPUs BY `rating` IN DESCENDING ORDER.
            4. SELECT THE HIGHEST-RATED GPU THAT MEETS ALL CRITERIA AND SAVE ITS DETAILS.

            #### STEP 2: SELECT A CPU
            1. QUERY THE `cpu_merged` TABLE TO RETRIEVE CPU OPTIONS.
            2. FILTER RESULTS BASED ON:
            - THE USER'S BUDGET USING `price <= user_budget`.
            - CHOOSE ONLY UNIQUE CPUS OR FOLLOW THE INSTRUCTIONS
            - ADDITIONAL REQUIREMENTS SUCH AS MINIMUM CORE COUNT (`core_count >= {required_cores}`).
            3. VERIFY COMPATIBILITY WITH THE MOTHERBOARD (SELECTED IN STEP 3) USING THE `socket_compatibility` TABLE.
            4. RANK THE CPUs BY `rating` IN DESCENDING ORDER.
            5. SELECT THE HIGHEST-RATED CPU THAT MEETS ALL CRITERIA AND SAVE ITS DETAILS.

            #### STEP 3: SELECT A MOTHERBOARD
            1. QUERY THE `motherboard` TABLE TO RETRIEVE OPTIONS.
            2. FILTER RESULTS BASED ON:
            - THE USER'S BUDGET.
            - CHOOSE ONLY UNIQUE MOTHERBOARDS OR FOLLOW THE INSTRUCTIONS
            - COMPATIBILITY WITH THE CPU (SELECTED IN STEP 2) USING THE `socket_compatibility` TABLE.
            3. VERIFY THAT THE MOTHERBOARD HAS SUFFICIENT:
            - MEMORY SLOTS (`memory_slots >= number_of_modules`).
            - MAXIMUM MEMORY (`max_memory >= required_memory`).
            4. SELECT THE MOST SUITABLE MOTHERBOARD WITHIN THE BUDGET AND SAVE ITS DETAILS.

            #### STEP 4: SELECT MEMORY (RAM)
            1. QUERY THE `memory` TABLE TO RETRIEVE RAM OPTIONS.
            2. FILTER RESULTS BASED ON:
            - THE USER'S BUDGET USING `price <= user_budget`.
            - SUFFICIENT MEMORY SLOTS ON THE SELECTED MOTHERBOARD (`memory_slots >= number_of_modules`).
            3. RANK RAM MODULES BY:
            - SPEED (HIGHER IS BETTER).
            - CAS LATENCY (LOWER IS BETTER).
            4. SELECT THE MOST SUITABLE RAM THAT MEETS ALL CRITERIA AND SAVE ITS DETAILS.

            #### STEP 5: SELECT A CASE
            1. QUERY THE `case` TABLE TO RETRIEVE OPTIONS.
            2. FILTER RESULTS BASED ON:
            - THE USER'S BUDGET.
            - COMPATIBILITY WITH THE MOTHERBOARD'S FORM FACTOR USING THE `case_motherboard_compatibility` TABLE.
            3. RANK CASES BY `price` IN ASCENDING ORDER.
            4. SELECT THE MOST SUITABLE CASE THAT MEETS ALL CRITERIA AND SAVE ITS DETAILS.

            ### WHAT NOT TO DO ###
            - NEVER IGNORE THE USER'S BUDGET CONSTRAINTS.
            - NEVER FAIL TO CONSIDER ADDITIONAL USER REQUIREMENTS.
            - NEVER SELECT COMPONENTS THAT ARE INCOMPATIBLE WITH EACH OTHER.
            - NEVER FAIL TO REFER TO THE COMPATIBILITY TABLES WHEN REQUIRED.
            - NEVER SELECT COMPONENTS ARBITRARILY WITHOUT RANKING BY RELEVANT CRITERIA.
            - NEVER SAVE INCOMPLETE OR INVALID COMPONENT DETAILS.

            ### EXAMPLES OF SQL QUERIES ###

            #### EXAMPLE 1: SELECTING A GPU WITH ADDITIONAL REQUIREMENTS
            ```sql
            SELECT *
            FROM gpu_full_info
            WHERE average_price <= {user_budget}
            AND memory >= {required_memory}
            AND EXISTS (
                SELECT 1
                FROM gpu_hierarchy
                WHERE gpu_hierarchy.gpu_id = gpu_full_info.id
                AND 4K_Ultra LIKE '%fps%'
            )
            ORDER BY rating DESC
            LIMIT 1;

                 """    
    )
    
    # Build the basic query
    query = f"Choose 2 unique {component} in the price range <= {budget}"
    
    if filters:
        filter_conditions = " AND ".join([f"{key} {value}" for key, value in filters.items()])
        query = f"Choose 2 unique {component} in the price range <= {budget} {filter_conditions} "
    
    # Execute the query using the agent and include the system prompt
    agent_result = agent_executor.invoke({
        "input": query,
        "system_prompt": system_prompt  # Add system prompt to guide the agent
    })
    
    # Convert the results into a list of Component objects
    data = agent_result.get('ouput', [])
    return [
        Component(
            name=row["name"],
            price=row["price"],
            additional_specs={k: row[k] for k in row if k not in ["name", "price"]}
        )
        for row in data
    ]

def generate_build(
    budget: float,
    build_type: str,
    components_percentages: Dict[str, Dict[str, float]],
    agent_executor,
    additional_filters: Optional[Dict[str, Dict[str, str]]] = None
) -> BuildResponse:
    """Generates a PC build based on the budget, type, and filters."""
    # Calculate the budget distribution across components
    budget_distribution = calculate_budget_distribution(budget, build_type, components_percentages)
    
    components = {}

    # Fetch components based on the budget for each category
    for component, component_budget in budget_distribution.items():
        filters = additional_filters.get(component) if additional_filters else None
        components[component] = fetch_components(agent_executor, component, component_budget, filters)
    
    return BuildResponse(
        type=build_type,
        budget=budget,
        components=components
    )


# Example budget and percentage breakdown for components
budget = 200000
build_type = "игровая"  # Could be other types like 'офисная', 'рабочая станция', etc.

components_percentages = {
    "игровая": {
        "gpu": 0.4,
        "cpu": 0.3,
        "memory": 0.1,
        "motherboard": 0.1,
        "power_supply": 0.05,
        "case": 0.05
    },
    "офисная": {
        "cpu": 0.4,
        "memory": 0.3,
        "motherboard": 0.2,
        "power_supply": 0.1
    }
}

# Example filters (optional)
additional_filters = {
    "gpu": {"memory": ">= 8 GB"},
    "cpu": {"core_count": ">= 6"}
}

# Generate the PC build
build_response = generate_build(budget, build_type, components_percentages, agent_executor, additional_filters)

# Display the result

````
